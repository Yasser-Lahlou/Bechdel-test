```python
import requests, re
from bs4 import BeautifulSoup
```

```python
URL = 'https://imsdb.com/all-scripts.html'
URL_BASE = 'https://www.imsdb.com/scripts/'
```

```python
def scrap_list(url):
    headers = {
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_0) \
                AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.146 Safari/537.36'
                }
    response = requests.get(url, headers=headers)
    with open('IMSDB_script_list', "w") as f:
        f.write(response.text)
```

```python
scrap_list(URL)
```

```python
def movie_list():
    with open('IMSDB_script_list', "r") as f:
        data = f.read()
    soup = BeautifulSoup(data, features="lxml")
    movies_list = soup.select('p a')
    list_urls = []
    for movie in movies_list:
        movie_title = movie['title']
        movie_title = movie_title.replace(' ', '-')
        movie_title = re.sub(r'\:', '', movie_title)
        movie_title = re.sub(r'-Script', '', movie_title)
        movie_url = URL_BASE + movie_title + '.html'
        list_urls.append([movie_title, movie_url])
    return list_urls
```

```python
def scrap_script(title, url):
    headers = {
                'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_0) \
                AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.146 Safari/537.36'
                }
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, features="lxml")
    tags = soup.select('pre')
    if len(tags) != 0:
        script = tags[-1]
        with open(f'scripts/{title}.txt', "w", encoding='utf-8', errors='ignore') as f:
            f.write(script.text)
    else:
        print(title)
```

```python
for movie in movie_list():
    scrap_script(movie[0], movie[1])
```
